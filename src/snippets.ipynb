### Extract apply link domains

The following code extract the domain of apply links. The extraction is used to decide which apply boards are mostly used. I'm aiming to automate the application process at these job boards.

__DON'T REMOVE__

```python
import pandas as pd
from src.db import DB
import re

db = DB("results/jobs.sqlite")
# https://something.domain.com/blahblahblah -> domain.com
pat = re.compile(r"https?:\/\/(.*\.)?(.*\..*?)\/.*")
q = """
SELECT d.id, d.apply_link, c.name FROM details as d 
LEFT JOIN company as c ON d.company_id = c.id
WHERE apply_link IS NOT NULL;
"""
db.cursor.execute(q)
raw = db.cursor.fetchall()
df = pd.DataFrame(columns=["id","apply_link","company_name"],data=raw)
df["apply_board"] = df["apply_link"].apply(lambda lnk: pat.findall(lnk)[0][-1])
df['apply_board'].value_counts()
```

```python
import subprocess
import os
from io import StringIO

def run_cpulimit(file_path:str="/home/ubuntu/cpulimit-all.sh",proc_name="chrome",cpu_pct=20):
    if not os.path.exists(file_path):
        print(f"cpu_limit file path does not exist. path:{file_path}")
        return None
    command = ["bash",file_path,"-l",str(cpu_pct),"-e",proc_name,"--max-depth=3","--watch-interval=2","-q"]
    try:
        proc = subprocess.Popen("ls",shell=False,stderr=subprocess.PIPE,stdout=subprocess.DEVNULL)
    except:
        # if the command is invalid or the file doesn't exist it throws an error
        return None
    # Check if we got errors from stderr
    if proc.stderr.!=b"":
        proc.kill()
        return None
    return proc

p =run_cpulimit("/home/arian/Desktop/cpulimit-all.sh")
if p == None:
    print("error!")
else:
    print("OK")
    p.kill()
```
